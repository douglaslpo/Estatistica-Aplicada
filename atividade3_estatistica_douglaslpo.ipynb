# **Questão 1. Teste de Normalidade**
* A distribuição abaixo apresenta os tempos, em segundos, que 75 computadores aleatoriamente
escolhidos por um cientista da computação gastaram para executar determinada tarefa.
  
  89, 93, 96, 98, 98, 101, 102, 102, 104, 105, 106, 108, 109, 109, 110,
110, 110, 111, 111, 112, 112, 113, 113, 114, 114, 114, 114, 114, 114, 114,
115, 115, 115, 116, 116, 116, 116, 116, 116, 116, 117, 117, 117, 118, 118,
118, 119, 119, 119, 119, 119, 119, 119, 120, 120, 120, 121, 121, 121, 122,
122, 123, 123, 124, 125, 125, 126, 127, 127, 127, 128, 128, 129, 131, 133.
* Discuta (ao nível de 5%) a normalidade desses dados.
* Faça um histograma e um boxplot dos dados. O resultado sobre a normalidade se alteraria
caso os outliers fossem desconsiderados?

````python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Dados fornecidos
data_questao1 = np.array([89, 93, 96, 98, 98, 101, 102, 102, 104, 105, 106, 108, 109, 109, 110,
                          110, 110, 111, 111, 112, 112, 113, 113, 114, 114, 114, 114, 114, 114, 114,
                          115, 115, 115, 116, 116, 116, 116, 116, 116, 116, 117, 117, 117, 118, 118,
                          118, 119, 119, 119, 119, 119, 119, 119, 120, 120, 120, 121, 121, 121, 122,
                          122, 123, 123, 124, 125, 125, 126, 127, 127, 127, 128, 128, 129, 131, 133])

# Teste de Normalidade (Shapiro-Wilk)
shapiro_stat, shapiro_p_value = stats.shapiro(data_questao1)
print(f"Teste de Normalidade (Shapiro-Wilk): Estatística = {shapiro_stat}, P-Valor = {shapiro_p_value}")

# Histograma
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.histplot(data_questao1, kde=True)
plt.title("Histograma dos Dados")

# Boxplot
plt.subplot(1, 2, 2)
sns.boxplot(x=data_questao1)
plt.title("Boxplot dos Dados")

plt.show()

# Análise sem outliers
# Vamos remover outliers usando o critério IQR (Intervalo Interquartil)
Q1 = np.percentile(data_questao1, 25)
Q3 = np.percentile(data_questao1, 75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

data_questao1_no_outliers = data_questao1[(data_questao1 >= lower_bound) & (data_questao1 <= upper_bound)]

# Teste de Normalidade sem outliers (Shapiro-Wilk)
shapiro_stat_no_outliers, shapiro_p_value_no_outliers = stats.shapiro(data_questao1_no_outliers)
print("\nTeste de Normalidade (sem outliers) - Shapiro-Wilk:",
      f"Estatística = {shapiro_stat_no_outliers}, P-Valor = {shapiro_p_value_no_outliers}")

# Histograma sem outliers
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.histplot(data_questao1_no_outliers, kde=True)
plt.title("Histograma dos Dados (sem outliers)")

# Boxplot sem outliers
plt.subplot(1, 2, 2)
sns.boxplot(x=data_questao1_no_outliers)
plt.title("Boxplot dos Dados (sem outliers)")

plt.show()
````

# **Questão 3. Teste para proporção em amostra única**

Quando Mendel realizou seus famosos experimentos sobre genética, uma amostra de descendentes
consistia em 428 ervilhas verdes e 152 ervilhas amarelas. Teste a um nível de significância de 5%, a
afirmativa de Mendel de que, sob as mesmas circunstâncias, 25% das ervilhas descendentes seriam
amarelas.

````python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Dados do problema
numero_ervilhas_amarelas = 152
numero_ervilhas_verdes = 428
proporcao_hipotetica = 0.25
nivel_significancia = 0.05

# Calcular a proporção observada
proporcao_observada = numero_ervilhas_amarelas / (numero_ervilhas_amarelas + numero_ervilhas_verdes)

# Realizar o teste de proporção
z_stat, p_value = sm.stats.proportions_ztest(numero_ervilhas_amarelas, numero_ervilhas_amarelas + numero_ervilhas_verdes, proporcao_hipotetica)

# Gráfico de barras
labels = ['Ervilhas Amarelas', 'Ervilhas Verdes']
observacoes = [numero_ervilhas_amarelas, numero_ervilhas_verdes]
proporcoes = [proporcao_observada, 1 - proporcao_observada]

fig, ax = plt.subplots()
ax.bar(labels, observacoes, color=['yellow', 'green'], alpha=0.7, label='Observado')
ax.bar(labels, [numero_ervilhas_amarelas, numero_ervilhas_verdes], color=['yellow', 'green'], alpha=0.3, label='Esperado', hatch='//')

# Adicionar barra de erro para a proporção hipotética
# O intervalo de confiança é obtido automaticamente pelo statsmodels
ax.errorbar(labels, [numero_ervilhas_amarelas * proporcao_hipotetica, numero_ervilhas_verdes * (1 - proporcao_hipotetica)],
            yerr=sm.stats.proportion_confint(numero_ervilhas_amarelas, numero_ervilhas_amarelas + numero_ervilhas_verdes),
            fmt='none', color='red', ecolor='red', capsize=5, label=f'Intervalo de Confiança ({1-nivel_significancia:.0%})')

plt.title('Teste de Proporção para Ervilhas Amarelas')
plt.ylabel('Número de Ervilhas')
plt.legend()
plt.show()

# Exibir estatísticas do teste
print(f'Estatística do teste Z: {z_stat}')
print(f'P-Valor: {p_value}')
````


A estatística do teste Z é uma medida que indica o quão longe a proporção observada está da proporção hipotética, levando em consideração a variabilidade esperada. No seu caso, a estatística do teste Z é aproximadamente 0.66.

O valor-p associado ao teste é 0.51, o que é maior que o nível de significância comumente escolhido de 0.05. O valor-p representa a probabilidade de observar uma estatística do teste tão extrema quanto a observada, assumindo que a hipótese nula é verdadeira. Quando o valor-p é alto, não temos evidências suficientes para rejeitar a hipótese nula.

Interpretação:

Estatística do Teste Z (0.66): A estatística do teste Z é uma medida padronizada que indica quantos desvios padrão a proporção observada está da proporção hipotética. Neste caso, 0.66 desvios padrão indicam uma discrepância moderada entre a observação e a expectativa sob a hipótese nula.

Valor-p (0.51): O valor-p é relativamente alto (maior que 0.05), o que sugere que a diferença entre a proporção observada e a proporção hipotética poderia ser devida ao acaso. Não temos evidências estatísticas convincentes para rejeitar a hipótese nula.

Conclusão para um Aluno de Mestrado:

Com base nos resultados do teste de proporção, não há evidências estatísticas suficientes para rejeitar a hipótese nula de que a proporção real de ervilhas amarelas é igual a 25%. Isso significa que, sob as circunstâncias do experimento, não podemos afirmar com confiança que a proporção de ervilhas amarelas é diferente da proporção proposta por Mendel.

É importante notar que, mesmo que a diferença observada não seja estatisticamente significativa, é sempre útil considerar o contexto do experimento, possíveis limitações e interpretações adicionais que podem ser relevantes para o campo da genética. Em alguns casos, resultados não significativos podem fornecer insights valiosos para pesquisas futuras.

# Questão 4. Teste para comparar médias de duas amostras pareadas

O gerente de uma empresa de táxi está tentando decidir se o uso de novo tipo de pneus aumenta a economia de combustível de sua frota. Doze carros foram equipados com os novos pneus e conduzidos por um curso de teste pré-estabelecido. Sem mudar de motorista, os mesmos carros foram então equipados com pneus dos mesmos tipos anteriores e conduzidos novamente durante o curso de teste.

O consumo de gasolina, em quilômetros por litro, foi registrado como segue:

| Carro | Km/L pneus tradicionais | Km/L pneus novos |
|-------|--------------------------|-------------------|
| 1     | 4,2                      | 4,1               |
| 2     | 4,7                      | 4,9               |
| 3     | 6,6                      | 6,2               |
| 4     | 7,0                      | 6,9               |
| 5     | 6,7                      | 6,8               |
| 6     | 4,5                      | 4,4               |
| 7     | 5,7                      | 5,7               |
| 8     | 6,0                      | 5,8               |
| 9     | 7,4                      | 6,9               |
| 10    | 4,9                      | 4,7               |
| 11    | 6,1                      | 6,0               |
| 12    | 5,2                      | 5,9               |

- Assumindo que as populações sejam normalmente distribuídas, pode-se concluir que carros equi-
pados com os novos pneus fornecem maior economia de combustível que os antigos? (Use 5% de
significância).


````python
import scipy.stats as stats

# Dados do problema
consumo_tradicionais = [4.2, 4.7, 6.6, 7.0, 6.7, 4.5, 5.7, 6.0, 7.4, 4.9, 6.1, 5.2]
consumo_novos = [4.1, 4.9, 6.2, 6.9, 6.8, 4.4, 5.7, 5.8, 6.9, 4.7, 6.0, 5.9]

# Realizar o teste t pareado
t_stat, p_value = stats.ttest_rel(consumo_tradicionais, consumo_novos)

# Exibir estatísticas do teste
print(f'Estatística do teste t: {t_stat}')
print(f'Valor-p: {p_value}')

# Decisão com base no valor-p
nivel_significancia = 0.05
if p_value < nivel_significancia:
    print('Rejeitar H0. Há evidências para afirmar que os novos pneus fornecem maior economia de combustível.')
else:
    print('Não rejeitar H0. Não há evidências para afirmar que os novos pneus fornecem maior economia de combustível.')
````

Estatística do Teste t (0.66): A estatística do teste t é uma medida padronizada que avalia a diferença entre as médias das duas amostras em relação à variabilidade dentro das amostras. Neste caso, a estatística do teste t é aproximadamente 0.66. Este valor indica que a diferença média no consumo de combustível entre os pneus tradicionais e os novos não é substancial em termos estatísticos.

Valor-p (0.52): O valor-p associado ao teste t é 0.52, que é maior que o nível de significância comumente escolhido de 0.05. Isso implica que não há evidências suficientes para rejeitar a hipótese nula. Em outras palavras, não podemos afirmar com confiança que os pneus novos proporcionam uma economia de combustível significativamente maior do que os pneus tradicionais.

Conclusão: Com base nos resultados do teste t, não encontramos evidências estatísticas convincentes para afirmar que os pneus novos oferecem uma economia de combustível significativamente maior do que os pneus tradicionais. O valor-p relativamente alto sugere que a diferença observada entre as médias pode ser devida ao acaso, e não há suporte estatístico para a afirmação de uma melhoria substancial na economia de combustível com os novos pneus.

Recomendações Adicionais: Como estudante de mestrado, seria apropriado discutir possíveis limitações do estudo, considerar a relevância prática da diferença observada, e sugerir análises adicionais ou pesquisas futuras para explorar mais a fundo as questões relacionadas ao desempenho dos pneus.

Valor-p (0.35): O valor-p associado ao teste t é 0.35, o que é maior que o nível de significância comumente escolhido de 0.05. Isso significa que não há evidências suficientes para rejeitar a hipótese nula. Em termos práticos, isso sugere que não podemos afirmar com confiança que os novos pneus proporcionam uma economia de combustível significativamente maior do que os antigos.

Conclusão: Com base nos resultados do teste t, não encontramos evidências estatísticas convincentes para afirmar que os novos pneus resultam em uma economia de combustível significativamente maior em comparação com os antigos. # Teste para Comparação de Médias de Duas Amostras Independentes

Em um experimento de adubação de pastagens, foram comparados dois tipos de adubação: química e orgânica. Seja Xq a altura das plantas adubadas com adubo químico e Xo a altura das adubadas com adubo orgânico, de modo que ambas tenham distribuição aproximadamente normal. Observaram-se 12 parcelas de Xq e 10 parcelas de Xo. Os resultados são apresentados a seguir:

```python
import numpy as np

# Alturas das plantas adubadas com adubo químico (Xq)
Xq = np.array([67.1, 56.4, 63.5, 60, 62.5, 59.6, 70, 67.9, 69.9, 74.1, 65.9, 58])

# Alturas das plantas adubadas com adubo orgânico (Xo)
Xo = np.array([82.7, 71.8, 70, 71.8, 78, 69.6, 75.8, 62.8, 72.4, 74.9])


````python
# -----------------Teste de Levene para Igualdade de Variâncias----------------------
# -Antes de realizar um teste t independente, é importante verificar se as          -
# -variâncias das duas amostras são iguais. Vamos usar um teste de Levene para isso.-
# -----------------------------------------------------------------------------------
from scipy.stats import levene

# Teste de Levene para igualdade de variâncias
statistic, p_value_levene = levene(Xq, Xo)
print(f'Estatística do teste de Levene: {statistic}')
print(f'P-valor do teste de Levene: {p_value_levene}')
````

````python
# -----------------------------Teste t Independente---------------------------------------
# -Se o p-valor do teste de Levene for maior que o nível de significância escolhido (5%),-
# -podemos prosseguir com o teste t independente.                                        -
# ----------------------------------------------------------------------------------------
from scipy.stats import ttest_ind

# Teste t independente
t_stat, p_value_ttest = ttest_ind(Xq, Xo, equal_var=(p_value_levene > 0.05))
print(f'Estatística do teste t: {t_stat}')
print(f'P-valor do teste t: {p_value_ttest}')
````

````python
import matplotlib.pyplot as plt
import numpy as np

# Alturas das plantas adubadas com adubo químico (Xq)
Xq = np.array([67.1, 56.4, 63.5, 60, 62.5, 59.6, 70, 67.9, 69.9, 74.1, 65.9, 58])

# Alturas das plantas adubadas com adubo orgânico (Xo)
Xo = np.array([82.7, 71.8, 70, 71.8, 78, 69.6, 75.8, 62.8, 72.4, 74.9])

# Criar um gráfico de barras para visualizar as alturas das plantas
fig, ax = plt.subplots(figsize=(20, 6))
ax.bar(np.arange(len(Xq)), Xq, width=0.5, label='Adubo Químico', align='center', color='blue')
ax.bar(np.arange(len(Xo)) + 0.6, Xo, width=0.5, label='Adubo Orgânico', align='center', color='orange')

# Configurar rótulos e título
ax.set_xticks(np.arange(len(Xq) + len(Xo)) / 2)
ax.set_xticklabels(['prc' + str(i + 1) for i in range(len(Xq) + len(Xo))])
ax.set_ylabel('Alturas das Plantas')
ax.set_title('Alturas das Plantas Adubadas com Adubo Químico e Orgânico')
ax.legend()

# Exibir o gráfico
plt.show()
````

## Interpretação:

**Estatística do Teste t:** O valor do \(t\)-statistic é -3.62, indicando uma diferença significativa entre as médias das alturas das plantas adubadas com adubo químico e orgânico.

**P-valor:** O valor-p associado ao teste t é 0.0017, o que representa a probabilidade de obter uma estatística de teste tão extrema quanto a observada, assumindo que não há diferença real entre as médias.

## Conclusão:

**Comparação de Médias:** Com base no teste t, não rejeitamos a hipótese nula de que não há diferença significativa entre as médias das alturas das plantas adubadas com adubo químico e orgânico (\(p > 0.05\)).

## Observações Adicionais:

Recomenda-se sempre verificar a igualdade de variâncias antes de realizar o teste t independente. Neste caso, utilizamos o Teste de Levene.
